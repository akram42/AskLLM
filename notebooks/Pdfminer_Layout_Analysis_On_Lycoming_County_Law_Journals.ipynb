# Pdfminer Layout Analysis

```python
pdf_path = "/content/drive/MyDrive/Lycoming County/682888_1124.pdf"
```

```python
# @title Display PDF
from pdf2image import convert_from_path
import matplotlib.pyplot as plt

def display_pdf(pdf_path):
    # Convert PDF pages to images
    images = convert_from_path(pdf_path)

    # Display each image
    for i, image in enumerate(images):
        plt.figure(figsize=(16, 16))
        plt.imshow(image)
        plt.axis('off')
        plt.title(f'Page {i+1}')
        plt.show()

# Display the PDF
display_pdf(pdf_path)
```

```python
# @title Pdfminer results
from pdfminer.high_level import extract_pages
from pdfminer.layout import LTTextBoxHorizontal, LTTextLineHorizontal, LTFigure, LTLine, LTCurve


for page_i, page in enumerate(extract_pages(pdf_path), start=1):
    page_header = f"{'=' * 100}\n{' ' * 40}Page {page_i}\n{'=' * 100}"
    print(page_header)

    for elm in page:
        print(type(elm).__name__)
        if isinstance(elm, (LTTextBoxHorizontal, LTTextLineHorizontal)):
            print(f'\nText: "{elm.get_text()}"')
        print('-' * 100)
```
